### -- Background --
Germany's professional league was formed in 1963 after the increased professionalisation of the sport overseas drew away many of the country's best players and Germany's faltering competitive status on the international stage. After a meeting at Dortmund, the professional league was formed by inviting the best-performing clubs across the lands.

### -- Formation --
By performance, the league has some discernable eras:
- Early 60s to mid 60s: Free for all.
- Late 60s to mid 70s: The struggle for supremacy between <b>Bayern Munich</b> and <b>Borussia Mönchengladbach</b>.
- Late 70s to early 80s: <b>1. FC Köln</b>, <b>Hamburger SV</b>, and <b>VfB Stuttgart</b> take centre stage.
- Mid 80s to late 80s: <b>Bayern Munich</b>'s return.
- Early 90s to mid 90s: <b>1. FC Kaiserslautern</b> and <b>Borussia Dortmund</b> claim their first titles.
- Late 90s to late 00s: <b>Bayern Munich</b> return to dominance.
- Since: <b>Bayern Munich</b> sustain dominance.

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
bundesligadata = pd.read_csv("bundesliga.csv", encoding='ISO-8859-1')
bundesligadata
```

That is almost everything we need, but there is still a few things missing. Let's do <b>Season</b>.

```python
season = bundesligadata['Year']+1
season = season.astype(str).str[2:4]
bundesligadata['Season'] = bundesligadata['Year'].astype(str)+'/'+season
bundesligadata
```

Now let's do <b>points per game</b> because not every season is exactly 34 matches and it used to be two points per win.

```python
bundesligadata['PPG'] = bundesligadata['Points (Winner)']/(bundesligadata['Wins']+bundesligadata['Draws']+bundesligadata['Losses'])
bundesligadata
```
