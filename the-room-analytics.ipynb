```python
from bs4 import BeautifulSoup as bs
import requests
```

The link used for scraping is done by a fan. Although unofficial, it's pretty accuracy:

https://medium.com/@DavidKlion/full-transcript-of-the-room-341e4286db8e

The reason the above link is used instead of the real script is because the movie doesn't actually follow the original script. As seen here:

https://www.scriptslug.com/assets/uploads/scripts/the-room-2003.pdf
    
A number of exchanges and scenes were either modified or moved around, or they were gone.

```python
r = requests.get("https://medium.com/@DavidKlion/full-transcript-of-the-room-341e4286db8e")
soup = bs(r.content)
contents = soup.prettify()
print(contents)
```

```python
info_box = soup.find(class_="ag ah ai aj ak fk am w")
info_rows = info_box.find_all("p")
for row in info_rows:
    print(row.prettify())
```

```python
script = {}

for index, row in enumerate(info_rows):
    script[index] = row.get_text(strip=True)
    
script
```

This gives us a dictionary with the script lines indexed. From there, we are going to split the script into dialogs and scenes.

```python
script_dialog = {}
script_scene = {}
dialog_index = 0
scene_index = 0

for index in script:
    if script[index].__contains__(':') and not script[index].__contains__('alarm'):
        dialog_index += 1
        script_dialog[dialog_index] = script[index]
    else:
        scene_index += 1
        script_scene[scene_index] = script[index]
```
