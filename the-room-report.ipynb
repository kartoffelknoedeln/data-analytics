## Generating a report for 'The Room'...

```python
from fpdf import FPDF

WIDTH = 210
HEIGHT = 297

def create_title_page(pdf):
    pdf.add_font('revue', '', r"c:\Users\Ken\AppData\Local\Microsoft\Windows\Fonts\Revue Regular.ttf", uni=True)
    pdf.set_font('revue', '', 48)
    pdf.cell(40, 20, 'THE ROOM')
    pdf.set_font('revue', '', 16)
    pdf.cell(0, 45, 'Directed, produced, and writted by Tommy Wiseau.')
    
    pdf.ln(5)
    pdf.image('theroomcover.jpg', 0, 50, WIDTH)
    
    pdf.ln(150)
    pdf.set_font('Arial', '', 11)
    pdf.cell(WIDTH, 50, 'A quick report over the infamous movie "The Room".')
    pdf.ln(5)
    pdf.cell(WIDTH, 50, 'Full analysis: https://github.com/kartoffelknoedeln/data-analytics/blob/master/the-room-report.ipynb')
    pdf.ln(20)
    pdf.cell(WIDTH, 55, 'Starring:')
    pdf.ln(3)
    pdf.set_font('Arial', 'B', 14)
    pdf.cell(WIDTH, 60, 'Tommy Wiseau')
    pdf.ln(3)
    pdf.cell(WIDTH, 65, 'Greg Sestero')
    pdf.ln(3)
    pdf.cell(WIDTH, 70, 'Juliette Danielle')
    
def create_overall_page(pdf):
    pdf.add_font('revue', '', r"c:\Users\Ken\AppData\Local\Microsoft\Windows\Fonts\Revue Regular.ttf", uni=True)
    pdf.set_font('revue', '', 16)
    pdf.cell(10, 20, 'Overall')
    
    pdf.ln(5)
    pdf.image('scenesandlines.png', 40, 30, WIDTH-50)
    
    pdf.ln(130)
    pdf.set_font('Arial', '', 11)
    pdf.cell(WIDTH, 50, 'The balance between lines and scenes in this movie is 80/20. Scenes taking up 20% of all lines is largely because')
    pdf.ln(5)
    pdf.cell(WIDTH, 50, 'the film has no method of character introduction. There are also random sex scenes that start the movie without')
    pdf.ln(5)
    pdf.cell(WIDTH, 50, 'explanation. For the rest of the script that has character dialog, many are inconsequential interactions with')
    pdf.ln(5)
    pdf.cell(WIDTH, 50, 'background characters, including many undeveloped subplots that were never mentioned again.')
    pdf.ln(20)
    pdf.cell(WIDTH, 50, 'There is also a sizeable amount of bizarre one-liners and quotes.')
    
def create_breakdown_page(pdf):
    pdf.add_font('revue', '', r"c:\Users\Ken\AppData\Local\Microsoft\Windows\Fonts\Revue Regular.ttf", uni=True)
    pdf.set_font('revue', '', 16)
    pdf.cell(10, 20, 'Breakdown')
    
    pdf.ln(5)
    pdf.image('lineseveryone.png', 60, 20, WIDTH-90)
    
    pdf.ln(85)
    pdf.set_font('Arial', '', 11)
    pdf.cell(WIDTH, 50, 'Johnny, Lisa, and Mark had 65% of the lines between themselves, and Lisa actually edges out the both of them')
    pdf.ln(5)
    pdf.cell(WIDTH, 50, "with the most lines. Of the other 35%, about 2/3's of it belonged to other minor characters such as Denny")
    pdf.ln(5)
    pdf.cell(WIDTH, 50, "Michelle (Lisa's friend), Claudette (Lisa's mother), and Peter (psychologist).")
    
    pdf.ln(30)
    pdf.image('scenery.png', 30, 140, WIDTH-45)
    
    pdf.ln(80)
    pdf.set_font('Arial', '', 11)
    pdf.cell(WIDTH, 50, '43% of the scenes were either people moving around, scenes of San Francisco, the football being thrown')
    pdf.ln(5)
    pdf.cell(WIDTH, 50, 'around, terrible flashback, and horrible sex scenes.')
    
def create_runtime_page(pdf):
    pdf.add_font('revue', '', r"c:\Users\Ken\AppData\Local\Microsoft\Windows\Fonts\Revue Regular.ttf", uni=True)
    pdf.set_font('revue', '', 16)
    pdf.cell(10, 20, 'Runtime')
    
    pdf.ln(5)
    pdf.image('runtime.png', 10, 25, WIDTH-10)
    
    pdf.ln(130)
    pdf.set_font('Arial', '', 11)
    pdf.cell(WIDTH, 50, 'Thanks to a YouTuber who broke the film down, only about 15% of the runtime contributes to the plot. Non-plot is')
    pdf.ln(5)
    pdf.cell(WIDTH, 50, 'nearly 3x the runtime and sitting at slightly above 41%. 7.5% is scenery, another 5% is yells and screams, and')
    pdf.ln(5)
    pdf.cell(WIDTH, 50, 'the final 30% or so are either dead time or scene progressions.')
    pdf.ln(20)
    pdf.cell(WIDTH, 50, 'That means at around 99 minutes, there is only 15 minutes of plot and over 40 minutes of subplots.')
    
def create_analytics_report(filename="the-room-report.pdf"):
    pdf = FPDF()
    
    '''Page 1: Cover'''
    pdf.add_page()
    create_title_page(pdf)
    
    '''Page 2: Overall'''
    pdf.add_page()
    create_overall_page(pdf)
    
    '''Page 3: Breakdown'''
    pdf.add_page()
    create_breakdown_page(pdf)
    
    '''Page 4: Runtime'''
    pdf.add_page()
    create_runtime_page(pdf)
    
    pdf.output('the-room-report.pdf', 'F')

if __name__ == '__main__':
    create_analytics_report()
```
